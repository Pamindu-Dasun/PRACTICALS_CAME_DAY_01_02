```python
raw_rows = [
    {'id': 'S001', 'label': 'cat', 'score': '0.95'},
    {'id': 'S002', 'label': 'dog', 'score': 'bad'},
    {'id': 'S003', 'label': 'cat', 'score': '0.82'},
    {'id': 'S004', 'label': 'bird', 'score': '0.61'}
]

valid_scores : list = []
label_counts : dict = {}
unique_labels : set = set()
```

```python
for raw in raw_rows:
    # get all valid scores and add it to the list[float]
    try:
        valid_scores.append(float(raw['score']))
    except ValueError:
        print(f'Skipped value {raw["score"]}')
    
    # calculate label count
    label = raw['label'].strip().lower()
    label_counts[label] = label_counts.get(label, 0) + 1

    unique_labels = set(label_counts.keys())
```

```python
print(f'Valid scores -> {valid_scores}')
print(f'Label counts -> {label_counts}')
print(f'Unique labels -> {unique_labels}')
```

```python
average_score = sum(valid_scores) / len(valid_scores)
print(f'Average score -> {average_score:.3f}')
```
