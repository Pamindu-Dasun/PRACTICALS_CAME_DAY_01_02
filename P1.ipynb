```python
sample_list = [1,2,3]
sample_tuple = (1,2,3)
sample_set = {1,2,3}

print(f'List : {sample_list}')
print(f'Tuple : {sample_tuple}')
print(f'Set : {sample_set}')
```

```python
# Ordered
print(f'First element of list -> {sample_list[0]}') # List is oredered data structure
print(f'First element of tuple -> {sample_tuple[0]}') # Tuple is ordered data structure

try:
    print(f'First element of set -> {sample_set[0]}')
except TypeError:
    print("set data structure is unordered.")
```

```python
# mutability
sample_list.append(4)
print(f'List after appending 4 -> {sample_list}')
```

```python
try:
    sample_tuple.append(4)
except AttributeError:
    print("Tuple is immutable data structure.")
```

```python
sample_set.add(4)
print(f'Set after adding 4 -> {sample_set}')
```

```python
# check duplicates
from random import randint

sample_list = [randint(1,50) for i in range (100)]
print(f'List -> {sample_list}')
print(f'Length of the list -> {len(sample_list)}')

sample_set : set = set(sample_list)
print(f'Set -> {sample_set}')
print(f'Length of the set -> {len(sample_set)}')
```

```python
# check duplicates count
from random import randint

sample_list = [randint(1,50) for i in range (100)]
print(f'List -> {sample_list}')
print(f'Length of the list -> {len(sample_list)}')

sample_set : set = set(sample_list)
print(f'Set -> {sample_set}')
print(f'Length of the set -> {len(sample_set)}')

sample_dict : dict = {}
for i in sample_list:
    sample_dict[i] = sample_dict.get(i,0) + 1

print(f'Duplicates count -> {sample_dict}')
```

```python
set_a = {1,3,5,7,9}
set_b = {1,2,3,4,5}

# common_set = set_a.intersection(set_b)
common_set = set_a & set_b
print(f'Common elements between set_a and set_b -> {common_set}')

print(f'Common set is subset of a set A -> {common_set.issubset(set_a)}')

all_set = set_a.union(set_b)
print(f'All elements in set_a and set_b -> {all_set}')
```
