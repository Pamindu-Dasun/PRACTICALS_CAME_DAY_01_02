## Question 1

**Topic:** _Core data types & None_ | **Marks:** 8 | **Difficulty:** Hard

You receive the following values from a broken CSV export:

`age` = '42', `income` = 0, `email` =", `is_active` = False, `notes` = None

Write a function is_training_ready(record) that returns True only when:
* age can be converted to an integer >= 18
* income is a number strictly greater than 0
* email is a non-empty string after .strip()
* is_active is True
* notes is None

Test your logic with the values above and explain in one comment why notes = None is different from email ="

***Starter data / skeleton:***

```python
age = '42'
income = 0
email = ''
is_active = False
notes = None
```

**Constraints**
```
· Do not use external libraries.
. Use is None for checking notes.
· Return a boolean only.
```

````python
def is_training_ready(record) -> bool:
    
    if record is None:
        return False
    elif not isinstance(record, dict):
        return False
    elif len(record) == 0:
        return False
    else:
        try:
            int(record['age'])
        except ValueError:
            return False
    
        try:
            float(record['income'])
        except ValueError:
            return False
        
        if record['email'].strip().lower() == '':
            return False
        
        if float(record['age']) < 18:
            return False
        
        if float(record['income']) < 0:
            return False
        
        if record['is_active'] == False:
            return False
        
        if record['notes'] is None:
            return False
        
        return True

````

````python
record = {
    'age': '42',
    'income': 0,
    'email': '',
    'is_active': False,
    'notes': None
}


is_training_ready(record)
````

Invalid labels must be skipped silently. Print the count of rejected labels.

    'positive',

]
```

**Constraints**

```


````python
raw_labels = [
    ' Positive ',
    'NEGATIVE',
    'neutral',
    'positive',
    ' Negative',
    'pos',
    ' negative '
]

clean_labels = []
````

### Question 5

**Topic:** _Boolean logic_ | **Marks:** 7 | **Difficulty:** Hard

A sample is eligible for model training only if all of the following are true:

- age is between 18 and 65 inclusive
- label is not 'spam'
- score is not None and score >= 0.60
- either country is 'LK' or the sample is marked is_premium = True
- has_missing_values is False

Write a function is_eligible(sample_dict) and determine how many samples in the dataset below are
eligible.

**Starter data / skeleton:**
```python
samples = [
{'age': 22, 'label': 'ham', 'score': 0.91, 'country': 'LK', 'is_premium': False, 'has_missing_values': False},
{'age': 17, 'label': 'ham', 'score': 0.91, 'country': 'LK', 'is_premium': False, 'has_missing_values': False},
{'age': 30, 'label': 'spam', 'score': 0.88, 'country': 'US', 'is_premium': True, 'has_missing_values': False},
{'age': 40, 'label': 'ham', 'score': None, 'country': 'US', 'is_premium': True, 'has_missing_values': False},
{'age': 50, 'label': 'ham', 'score': 0.75, 'country': 'US', 'is_premium': False, 'has_missing_values': False},
]
```

**Constraints**
```
. Use and, or, and not clearly.
. Check score with 'is not None' before comparing.
```

````python
def is_eligible(sample_dict: dict) -> bool:
    
    age = sample_dict.get('age')
    label = sample_dict.get('label')
    score = sample_dict.get('score')
    country = sample_dict.get('country')
    is_premium = sample_dict.get('is_premium')
    has_missing_values = sample_dict.get('has_missing_values')

    if age < 18 or age > 66:
        return False
    
    if label == 'spam':
        return False
    
    if score is None or score < 0.6:
        return False
    
    if country != 'LK' and is_premium is False:
        return False
    
    if has_missing_values is True:
        return False

    return True
````

````python
samples = [
    {'age': 22, 'label': 'ham', 'score': 0.91, 'country': 'LK',
        'is_premium': False, 'has_missing_values': False},
    {'age': 17, 'label': 'ham', 'score': 0.91, 'country': 'LK',
        'is_premium': False, 'has_missing_values': False},
    {'age': 30, 'label': 'spam', 'score': 0.88, 'country': 'US',
        'is_premium': True, 'has_missing_values': False},
    {'age': 40, 'label': 'ham', 'score': None, 'country': 'US',
        'is_premium': True, 'has_missing_values': False},
    {'age': 50, 'label': 'ham', 'score': 0.75, 'country': 'US',
        'is_premium': False, 'has_missing_values': False},
]

eligible_count = 0
for sample in samples:
    if is_eligible(sample):
        eligible_count += 1

print(f"Eligible samples count: {eligible_count}")
````

### Question 6

**Topic:** _Loops, break & continue_ | **Marks:** 8 | **Difficulty:** Hard

Process the scores list below. Build valid_scores containing only numeric values. Rules:

* Skip None using continue
* If any score is strictly less than 0.30, stop processing entirely using break and print 'Critical anomaly detected'
* Otherwise append valid scores

After the loop, print valid_scores and the number of processed items before stopping.

**Starter data / skeleton:**
```python
scores = [0.91, 0.88, None, 0.95, 0.25, 0.72, 0.81]
```

**Constraints**
```
· Use a for loop.
· Do not use try/except in this question.
```

````python
scores = [0.91, 0.88, None, 0.95, 0.25, 0.72, 0.81]
valid_scores = []

processed_item = 0

for score in scores:

    processed_item += 1

    if score is None:
        continue
    elif score < 0.30: 
        print('Critical anomaly detected')
        break
    else:
        valid_scores.append(score)

print(f'Valid scores -> {valid_scores}')
print(f'Processed items -> {processed_item}')
````

### Question 7

**Topic:** _while loops_ | **Marks:** 7 | **Difficulty:** Medium-Hard

A training job retries until accuracy reaches at least 0.90 or 5 attempts are used. Simulated accuracies are stored in attempt_results.

Write a while loop that prints:
Attempt 1: accuracy=0.72
Attempt 2: accuracy=0.81
...

Stop early if threshold is met. At the end, print whether training succeeded.

**Starter data / skeleton:**
```python
attempt_results = [0.72, 0.81, 0.86, 0.93, 0.94]
threshold = 0.90
max_attempts = 5
```

**Constraints**
```
· Use a while loop, not a for loop.
· Use attempt numbering starting from 1.
```

````python
attempt_results = [0.72, 0.81, 0.86, 0.93, 0.94]
threshold = 0.90
max_attempts = 5

current_attempt = 0

while current_attempt < max_attempts and current_attempt < len(attempt_results):
    print(f'Attempt {current_attempt + 1}: accuracy={attempt_results[current_attempt]}')
    
    if attempt_results[current_attempt] >= threshold:
        print('Training succeeded')
        break

    current_attempt += 1
````

### Question 8

**Topic:** _Exception handling_ | **Marks:** 8 | **Difficulty:** Hard

You are cleaning confidence values collected from a web form. Write code that:

* Converts each raw value to float when possible
* Skips invalid values with a message: Skipping invalid value: <value>
* Ignores empty strings after stripping
* Stores valid floats in clean_values
* Prints the final list and its average rounded to 3 decimal places

If clean_values is empty, print 'No valid values' instead of dividing by zero.

**Starter data / skeleton:**
```python
raw_values = ['0.91', 'bad', '   ', '0.72', '0.88x', '0.65', '']
```

**Constraints**
```
· Catch only ValueError.
· Do not use bare except.
```

````python
raw_values = ['0.91', 'bad', '   ', '0.72', '0.88x', '0.65', '']
clean_values = []

for raw_value in raw_values:

    if raw_value.strip() == '':
        continue

    try:
        float_value = float(raw_value)
        clean_values.append(float_value)
    except ValueError:
        print(f'Skipped invalid value: {raw_value}')


print(f'Final clean values: {clean_values}')
print(f'Average {round(sum(clean_values) / len(clean_values) ,3)}')
````

### Question 10

**Topic:** _Lists & processing_ | **Marks:** 8 | **Difficulty:** Hard

Given predictions and true_labels of equal length, build a list called mistakes containing tuples of (index, predicted, actual) for every wrong prediction. Then print the mistake count and the mistake with the highest index.

**Starter data / skeleton:**
```python
predictions = ['cat', 'dog', 'cat', 'bird', 'dog']
true_labels = ['cat', 'cat', 'cat', 'bird', 'cat']
```

**Constraints**
```
· Use enumerate().
· Use tuple values in mistakes.
```

````python
predictions = ['cat', 'dog', 'cat', 'bird', 'dog']
true_labels = ['cat', 'cat', 'cat', 'bird', 'cat']

mistakes = []

for index, predicted in enumerate(predictions):

    if predicted != true_labels[index]:
        mistakes.append((index, predicted, true_labels[index]))
    
print(f'Mistakes count : {len(mistakes)}')
print(f'Mistakes with the highest index : {mistakes[-1][0]}')
    
````

### Question 11

**Topic:** _Tuples_ | **Marks:** 6 | **Difficulty:** Medium-Hard

Each model result is stored as (label, confidence). Write code that unpacks each result and prints only those with confidence >= 0.85 in this format:
cat -> 0.97

Then create a tuple named best_result containing the label and confidence of the highest-confidence prediction.

**Starter data / skeleton:**
```python
results = [('cat', 0.97), ('dog', 0.62), ('bird', 0.88), ('cat', 0.75)]
```

**Constraints**
```
· Use tuple unpacking in the loop.
· Do not convert results to a dictionary.
```

````python
results = [('cat', 0.97), ('dog', 0.62), ('bird', 0.88), ('cat', 0.75)]

best_result = None

for label, confidence in results:

    if confidence >= 0.85:
        print(f'{label} -> {confidence}')
    
    if best_result is None:
        best_result = label, confidence
    else:
        if confidence > best_result[-1]:
            best_result = label, confidence

print(f'Best result: {best_result[0]} with confidence {best_result[1]}')
````

### Question 12

**Topic:** _Sets_ | **Marks:** 7 | **Difficulty:** Hard

train_labels and test_labels are provided. Find:

* labels that appear in test but not in train
* labels common to both splits
* whether test_labels contains duplicate values

Store results in variables and print them with clear headings.

**Starter data / skeleton:**
```python
train_labels = ['cat', 'dog', 'bird', 'cat', 'dog']
test_labels = ['dog', 'fish', 'cat', 'fish', 'rabbit']
```



````python
train_labels = ['cat', 'dog', 'bird', 'cat', 'dog']
test_labels = ['dog', 'fish', 'cat', 'fish', 'rabbit']

train_labels_set = set(train_labels)
test_labels_set = set(test_labels)

labels_only_in_test = test_labels_set.difference(train_labels_set)
print(f'Labels only in test: {labels_only_in_test}')

labels_common_in_both = test_labels_set.intersection(train_labels_set)
print(f'Labels common in both: {labels_common_in_both}')

print(f"Whether Test labels contains duplicate values {'YES' if len(test_labels) != len(test_labels_set) else 'NO'}")
````
