## Question 12
**Topic:** Sets | **Marks:** 7 | **Difficulty:** Hard

`train_labels` and `test_labels` are provided. Find:
* labels that appear in test but not in train
* labels common to both splits
* whether `test_labels` contains duplicate values

Store results in variables and print them with clear headings.

**Starter data / skeleton:**

```python
train_labels = ['cat', 'dog', 'bird', 'cat', 'dog']
test_labels = ['dog', 'fish', 'cat', 'fish', 'rabbit']

```

````python
train_labels = ['cat', 'dog', 'bird', 'cat', 'dog']
test_labels = ['dog', 'fish', 'cat', 'fish', 'rabbit']

train_labels_set = set(train_labels)
test_labels_set = set(test_labels)

print(f'labels that appear in test but not in train -> {test_labels_set.difference(train_labels_set)}')

print(f'labels common to both splits -> {test_labels_set.intersection(train_labels_set)}')

print(f'whether test_labels contains duplicate values -> {len(test_labels) != len(test_labels_set)}')
````
